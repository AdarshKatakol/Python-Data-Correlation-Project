```python
# Importing libraries

import pandas as pd
import seaborn as sns
import numpy as np
#! pip install matplotlib

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
#from matplot.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjusts the configuration of the plots we will create

# read the data

df = pd.read_csv('C:\College\Projects\Python Data correlation\movies.csv')
```

```python
# Looking at the data

df.head()
```

```python
# Data cleaning 

# View data types of all columns

df.dtypes
```

```python
# changing the data types of the coulmns "budget" and "gross"

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Fixing the "Released" and the "year" coulmn

df['released'] = df['released'].astype(str).str[:4]
df
```

```python
# Order our Data in decending order

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python

df.drop_duplicates()
```

```python
# Ranking companies based on average revenues generated using groupby()

df_test = df[['company', 'gross']]
df_grp = df_test.groupby(['company'], as_index= False).mean()
#df_grp.sort_values(by=['gross'], inplace=False, ascending=False)
df_grp

```

```python
df_grp.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Looking at the top 15 compaies by gross revenue

df_group_sum = df.groupby('company')[["gross"]].sum()

df_group_sum_Sorted = df_group_sum.sort_values('gross', ascending = False)[:15]

df_group_sum_Sorted = df_group_sum_Sorted['gross'].astype('int64') 

df_group_sum_Sorted
```

```python
# sum of gross revenue made by top 15 countries by year

df_Gross_Sum = df.groupby(['company', 'year'])[["gross"]].sum()

df_Gross_Sum_Sorted = df_Gross_Sum.sort_values(['gross','company','year'], ascending = False)[:15]

df_Gross_Sum_Sorted = df_Gross_Sum_Sorted['gross'].astype('int64') 

df_Gross_Sum_Sorted
```

```python
# Correlation between budget and gross revenue

# Plot budget vs gross using seaborn

sns.regplot (x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "red"}, line_kws = {"color":"black"})
```

```python
# showing correlation between the numeric features

df.corr()
```

```python
# Heat map of correlation between the numeric features

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# converting non-numeric columns into numeric columns

df_numerized = df.apply(lambda x: x.factorize()[0]).corr(method='spearman') 

#factorize assigns a random numeric value for each unique categorical value

df_numerized
```

```python
# Heat map of correlation between the all the features

complete_correlation_matrix = df_numerized.corr(method='spearman')

sns.heatmap(complete_correlation_matrix, annot = True)

plt.title("Correlation matrix of all the Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#unstacking 

corr_pairs = complete_correlation_matrix.unstack()
corr_pairs
```

```python
#F actors that have a high correlation (> 0.5)

sorted_pairs = corr_pairs.sort_values()

#strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

strong_pairs
```
